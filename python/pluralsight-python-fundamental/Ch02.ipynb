## 2.3 Significant whitespace in Python

(1) Indent 4 spaces by each level.

````python
for i in range(5):
    x = i * 10
    print(x)
    
````

(2) Advantages of "significant whitespace"

* Requires readable code: uniform style of indentation
* No clutter
* Human and computer can't get out of sync.

(3) Significant whitespace rules

"Programming as Guido indented it"

* Prefer four spaces to tabs.
* Never mix spaces and tabs.
* Be consistent on consecutive lines.
* Only deviate to improve readability: e.g., break long statements over multiple lines.

## 2.9 While loops

Two types of loops: for and while

```python
while expr:
    print("loop while expr is True")
```

"Explicit is better than implicit"

````python
c = 5
# Explicit comparison with zero -- Good!
while c != 0:
    print(c)
    c -= 1
    
````

````python
c = 5
# Implicit comparison with zero -- Bad!
while c:
    print(c)
    c -= 1
    
````

````python
# Infinite loop
while True:
    print("Looping!")
````

```python
# No do-while in python
```
