## 5.2 Demo: Starting a debugging session

(1) Start a pdb session in the code.

```python
import pdb
pdb.set_trace()
```

* In the pdb session, you can evaluate any Python code in the context of the running code.

* `l`: List ten lines of code centered at the running line.

## 5.3 Demo: Simple debuggin commands

More pdb commands:

(1) `n`: Go to next statement.

(2) `s`: Step into the function.

(3) Use "arrow up" key to recall previous commands.

## 5.4 Review: Simple debugging commands

Interactive use

(1) Loading and starting the debugger

```python
import pdb
pdb.set_trace()
```

| Key | Mnemonic  | Function                   |
| ----|:---------:| --------------------------:|
| `l` | list      | Show current lines         |
| `n` | next      | Execute line, move to next |
| `s` | step into | Step into the function     |
| `h` | help      | Provides documentation     |

(2) Use up and down arrows for history.

## 5.5 Demo: Advanced debugging commands

(1) `w(here)`: Print our the current stack.

(2) `b(reak)`: Set a breakpoint by using either the line number or the method name.

(3) `c(ontinue)`

(4) `q(uit)`: Quit the debugger.

## 5.6 Review: Advanced debugging commands

| Key | Mnemonic   | Function                     |
| ----|:----------:| ----------------------------:|
| `c` | continue   | Continue normal program flow |
| `b` | breakpoint | Set breakpoint               |
| `w` | where      | Show stacktrace              |
| `s` | step       | Step down into next level    |
| `r` | return     | Step out of this level       |
